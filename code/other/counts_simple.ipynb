```python
import pandas as pd
import numpy as np
import os
import collections
```

```python
files = [os.path.join(dp, f) for dp, dn, fn in os.walk(os.path.expanduser("./")) for f in fn if ".csv" in f]
```

```python
word_count = collections.defaultdict(int)
nan_count = 0
comment_count = collections.defaultdict(int)
for _ in files:
    df = pd.read_csv(_)
    for user in df['Username']:
        comment_count[user] += 1
    if df["Message Content"].dtype == object:
        for comment in list(df['Message Content'].str.split(" ")):
            if comment is not np.nan:
                for word in comment:
                    word_count[word] += 1
            else:
                nan_count += 1
```

```python
sorted_word = {k: v for k, v in sorted(word_count.items(), key=lambda item: item[1], reverse=True)}
```
